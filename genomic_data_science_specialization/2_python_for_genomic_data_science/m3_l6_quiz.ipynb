```python
import datetime as dt

def count1(dna, base):
    i = 0
    for c in dna:
        if c == base:
            i += 1 
    return i

def count2(dna, base):
    i = 0 
    for j in range(len(dna)):
        if dna[j] == base:
            i += 1 
    return i 

def count3(dna, base):
    match = [c == base for c in dna]
    return sum(match)

def count4(dna, base):
    return dna.count(base)

def count5(dna, base):
    return len([i for i in range(len(dna)) if dna[i] == base])

def count6(dna,base):
    return sum(c == base for c in dna)
```

```python
base = 'c'
dna = create_dna(1000000)
```

```python
start = dt.datetime.now()
print(count1(dna, base))
print((dt.datetime.now()-start).microseconds / 1000)
print()
```

```python
start = dt.datetime.now()
print(count2(dna, base))
print((dt.datetime.now()-start).microseconds / 1000)
print()
```

```python
start = dt.datetime.now()
print(count3(dna, base))
print((dt.datetime.now()-start).microseconds / 1000)
print()
```
